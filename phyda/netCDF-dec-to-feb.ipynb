```python
import netCDF4
import numpy as np
import glob
import os
```

```python
file = "da_hydro_DecFeb_r.1-2000_d.05-Jan-2018.nc"
f = netCDF4.Dataset(file)
```

```python
MIN_YEAR = 1630
MAX_YEAR = 1690
MIN_LAT = -40
MAX_LAT = 50
MIN_LON = 13
MAX_LON = 155
```

```python
columns = "ID,Date,Latitude,Longitude,pdsi_mn,pdsi_sg,pdsi_pc_5th,pdsi_pc_50th,pdsi_pc_95th,spei_mn,spei_sg,spei_pc_5th,spei_pc_50th,spei_pc_95th\n"
```

```python
id = 131272
year_indices = []
lats_dict = {}
lons_dict = {}

for i in range(MIN_YEAR-1, MAX_YEAR):
    year_indices.append(i)

# get a desired range of latitudes
for i in range(len(lat)):
    if lat[i] >= MIN_LAT and lat[i] <= MAX_LAT:
        lats_dict[i] = float(lat[i].data)

# get a desired range of longitudes
for i in range(len(lon)):
    if lon[i] >= MIN_LON and lon[i] <= MAX_LON:
        lons_dict[i] = float(lon[i].data)
        
all_lats = list(lats_dict.keys())
all_lons = list(lons_dict.keys())
```

```python
for year in year_indices[49:]:
    
    file_name = "/Users/Sienna/Desktop/files/" + str(year+1) + ".csv"
    f_cur = open(file_name, "w")
    f_cur.write(columns)
    
    for lat in all_lats:
        for lon in all_lons:
            
            id_out = str(id)
            date = str(year + 1) + "-01" + "-01"
            latitude = str(lats_dict[lat])
            longitude = str(lons_dict[lon])
                    
            if isinstance(pdsi_mn[year][lat][lon], np.ma.core.MaskedConstant):
                pdsi_mn_out = "ND"
                pdsi_sg_out = "ND"
                pdsi_pc5_out = "ND"
                pdsi_pc50_out = "ND"
                pdsi_pc95_out = "ND"
                spei_mn_out = "ND"
                spei_sg_out = "ND"
                spei_pc5_out = "ND"
                spei_pc50_out = "ND"
                spei_pc95_out = "ND"
            else:
                pdsi_mn_out = str(float(pdsi_mn[year][lat][lon]))
                pdsi_sg_out = str(float(pdsi_sg[year][lat][lon]))
                pdsi_pc5_out = str(float(pdsi_pc[0][year][lat][lon]))
                pdsi_pc50_out = str(float(pdsi_pc[1][year][lat][lon]))
                pdsi_pc95_out = str(float(pdsi_pc[2][year][lat][lon]))
                spei_mn_out = str(float(spei_mn[year][lat][lon]))
                spei_sg_out = str(float(spei_sg[year][lat][lon]))
                spei_pc5_out = str(float(spei_pc[0][year][lat][lon]))
                spei_pc50_out = str(float(spei_pc[1][year][lat][lon]))
                spei_pc95_out = str(float(spei_pc[2][year][lat][lon]))

            
            to_write = id_out + "," + date + "," + latitude + "," + longitude + "," + pdsi_mn_out \
            + "," + pdsi_sg_out + "," + pdsi_pc5_out + "," + pdsi_pc50_out + "," + pdsi_pc95_out + "," + \
            spei_mn_out + "," + spei_sg_out + "," + spei_pc5_out + "," + spei_pc50_out + "," + spei_pc95_out + "\n"
            
            f_cur.write(to_write)

            id += 1
    f_cur.close()
```

```python
f_final = open("/Users/Sienna/Desktop/dec_to_feb.csv", "w")
f_final.write(columns)

filelist = glob.glob(os.path.join('/Users/Sienna/Desktop/files/', '*.csv'))
for infile in sorted(filelist): 
    f = open(infile, "r")
    lines = f.readlines()
    for line in lines[1:]:
        f_final.write(line)
    f.close()

f_final.close()
```
